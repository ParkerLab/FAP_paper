```r
library(monocle3)
library(SeuratWrappers)

# The tutorial shown below and on subsequent pages uses two additional packages:
library(ggplot2)
library(dplyr)
```

```r
#first getting my seurat object loaded
setwd("~/links/christav/fap_village_multiome/results/RNA_clustering/monocle/")
rna <- readRDS("../RNA_final.rds")
```

```r
cds <- as.cell_data_set(rna)
```

```r
cds <- cluster_cells(cds, resolution=1e-5)
plot_cells(cds)
```

```r
cds <- learn_graph(cds)
```

```r
plot_cells(cds,
           label_groups_by_cluster=FALSE,
           label_leaves=FALSE,
           label_branch_points=TRUE,
          label_principal_points=TRUE)
```

```r
jpeg(file="rna_final_monocle_traj.jpeg", res=600, width=4800, height=4800)
plot_cells(cds,
           label_groups_by_cluster=TRUE,
           label_leaves=FALSE,
           label_branch_points=FALSE)
dev.off()
```

```r
cds <- order_cells(cds, root_pr_nodes='Y_8') #where I think progenitors are, left side
```

```r
plot_cells(cds,
           color_cells_by = "pseudotime",
           label_cell_groups=FALSE,
           label_leaves=FALSE,
           label_branch_points=FALSE,
           graph_label_size=1.5)
```

```r
jpeg(file="rna_final_monocle_pseudotime.jpeg", res=600, width=4800, height=4800)
cds <- order_cells(cds, root_pr_nodes='Y_8')
plot_cells(cds,
           color_cells_by = "pseudotime",
           label_cell_groups=FALSE,
           label_leaves=FALSE,
           label_branch_points=FALSE,
           graph_label_size=1.5)
dev.off()
```
