```r
library("org.Hs.eg.db")
```

```r
gseKegg <- function(sorted_genes) { # `genes` has to be sorted
    hs <- org.Hs.eg.db
    my.symbols <- names(sorted_genes)
    g <- AnnotationDbi::select(hs, keys = my.symbols, columns = c("ENTREZID", "SYMBOL", "ENSEMBL"), keytype = "ENSEMBL")
    g <- g[!is.na(g$ENTREZID),]
    g <- g[!duplicated(g$ENSEMBL),]
    sorted_genes <- sorted_genes[names(sorted_genes) %in% g$ENSEMBL]
    names(sorted_genes) <- g$ENTREZID
    res <- gseKEGG(sorted_genes, organism = "hsa", keyType = "kegg", pvalueCutoff = 1)
    
    return(res)
}

gseGo <- function(sorted_genes) { # `genes` has to be sorted
    res <- gseGO(sorted_genes, OrgDb=org.Hs.eg.db, ont = "BP", pvalueCutoff = 1, keyType = "ENSEMBL")
    return(res)
}

genes <- read.table("/scratch/scjp_root/scjp1/christav/fap_village_multiome/results/mast/all.fcHurdle.final.txt", header = T)
genes$ID <- sub("\\..*", "", rownames(genes))

tmp <- genes[!is.na(genes$fdr),]
tmp <- tmp[order(tmp$log2FC, decreasing = T),]
genelist <- tmp$log2FC
names(genelist) <- tmp$ID
gsea_go <- gseGo(genelist)

gsea_kegg <- gseKegg(genelist)
```

```r
head(gsea_go)
write.table(gsea_go, "gsea_go.final.txt", sep = "\t", col.names = TRUE, row.names = FALSE, quote = FALSE)
```

```r
head(gsea_kegg)
write.table(gsea_kegg, "gsea_kegg.final.txt", sep = "\t", col.names = TRUE, row.names = FALSE, quote = FALSE)
```
